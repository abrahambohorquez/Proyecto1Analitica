```python
# Pivot para calcular diferencias fácilmente
pivot = df.pivot_table(
    index='cole_area_ubicacion',
    columns='cole_naturaleza',
    values='punt_global',
    aggfunc='mean'
)

pivot['brecha_no_oficial_menos_oficial'] = pivot['NO OFICIAL'] - pivot['OFICIAL']

pivot
```

Ho: Media rural= Media urbano
H1: Media rural =/ Media urbano

```python
from scipy.stats import ttest_ind
import numpy as np

# Separar grupos
grupo1 = df[df['cole_area_ubicacion']=='RURAL']['punt_global']
grupo2 = df[df['cole_area_ubicacion']=='URBANO']['punt_global']

# Prueba t
t_stat, p_val = ttest_ind(grupo1, grupo2, nan_policy='omit')

print("RESULTADOS PRUEBA T")
print("---------------------")
print(f"Media Rural: {np.mean(grupo1):.2f}")
print(f"Media Urbano: {np.mean(grupo2):.2f}")
print(f"Diferencia de medias: {np.mean(grupo2) - np.mean(grupo1):.2f}")
print(f"Estadístico t: {t_stat:.2f}")
print(f"Valor p: {p_val:.10f}")

```

Ho: Media oficial = Media no oficial
H1: Media oficial =/ Media no oficial

```python
from scipy.stats import ttest_ind
import numpy as np

# Separar grupos
oficial = df[df['cole_naturaleza']=='OFICIAL']['punt_global']
no_oficial = df[df['cole_naturaleza']=='NO OFICIAL']['punt_global']

# Prueba t
t_stat, p_val = ttest_ind(oficial, no_oficial, nan_policy='omit')

print("RESULTADOS PRUEBA T - NATURALEZA DEL COLEGIO")
print("---------------------------------------------")
print(f"Media Oficial: {np.mean(oficial):.2f}")
print(f"Media No Oficial: {np.mean(no_oficial):.2f}")
print(f"Diferencia (No Oficial - Oficial): {np.mean(no_oficial) - np.mean(oficial):.2f}")
print(f"Estadístico t: {t_stat:.2f}")
print(f"Valor p: {p_val:.10f}")
```
